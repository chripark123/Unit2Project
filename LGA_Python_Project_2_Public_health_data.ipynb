```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv")
df.head()
```

```python
#1 Cleaning Task
#(replace ? with NaN)

df = df.replace('?', pd.NA)
```

```python
#2 Compare Groups
#Male vs. Female Patients on Number of Inpatient visits

df_male = df[df['gender'] == 'Male']
df_female = df[df['gender'] == 'Female']

inpatient_visits_male = df_male['number_inpatient'].sum()
inpatient_visits_female = df_female['number_inpatient'].sum()

print("Number of Inpatient Visits for Males:", inpatient_visits_male)
print("Number of Inpatient Visits for Females:", inpatient_visits_female)
```

```python
#3 Descriptive Statistics

print("Male Patients - Inpatient Visits Stats")
print("Average:", df_male['number_inpatient'].mean())
print("Min:", df_male['number_inpatient'].min())
print("Max:", df_male['number_inpatient'].max())

print("\nFemale Patients - Inpatient Visits Stats")
print("Average:", df_female['number_inpatient'].mean())
print("Min:", df_female['number_inpatient'].min())
print("Max:", df_female['number_inpatient'].max())
```

```python
#4 Visualization

# Set up data
genders = ['Male', 'Female']
totals = [inpatient_visits_male, inpatient_visits_female]

# Plot
plt.bar(genders, totals, color=['blue', 'pink'])
plt.title('Total Inpatient Visits by Gender')
plt.ylabel('Total Visits')
plt.xlabel('Gender')
plt.show()
```

#Conclusion

The total number of inpatient visits differs slightly between Male an Female patients. While both groups have similar average visits, small differences in total visits might reflect population differences or access to care.
